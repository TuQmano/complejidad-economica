```python
import pandas as pd
import csv

def csvToArrayOfDictionaries(csv):
  result = []
  for x in csv:
    countRow = 0;
    for y in csv[x]:
      if(len(result) < (countRow+1)): 
        result.append({})

      result[countRow][x] = y
      countRow+=1

  return result

clasif = csvToArrayOfDictionaries(pd.read_csv('/content/DESCRIPCION PARTIDAS Y SUBPARTIDAS - SUBPARTIDAS (6 DIGITOS).csv'))
green_products = csvToArrayOfDictionaries(pd.read_csv('/content/green_products-old.csv'))
nodes = csvToArrayOfDictionaries(pd.read_csv('/content/informacion_nodos_2014_2018 - informacion_nodos_2014_2018.csv'))
prod30 = csvToArrayOfDictionaries(pd.read_csv('/content/Tabla 30 Productos.csv'))
environmental_services = csvToArrayOfDictionaries(pd.read_csv('/content/servicios.csv'))

def findClasification(x):
  try:
    for c in clasif:
      if(int(c["SubPartidas Arancelarias"]) == x):
        return c["Descripción del producto"]
  except ValueError:
    return ''
    
  return ''

def findCategory(x):
  for n in nodes:
    if(n["prod_code"] == x):
      return n["category_description"]

def findSelectionCriteria(x):
  for n in prod30:
    if(n["HS92"] == x):
       if(n["Criterio de selección"] == 'Proximidad y complejidad'):
         return 'proximidad|complejidad'
       if(n["Criterio de selección"] == 'Proximidad y complejidad; VCR y complejidad'):
         return 'proximidad|complejidad|VCR'
       if(n["Criterio de selección"] == 'Proximidad y complejidad; VCR 2011 e ICP.'):
         return 'proximidad|complejidad|VCR2011|ICP'
       if(n["Criterio de selección"] == 'VCR y complejidad'):
         return 'VCR|complejidad'
       if(n["Criterio de selección"] == 'VCR 2011 e ICP'):
         return 'VCR|ICP'
       if(n["Criterio de selección"] == 'VCR 2011 e ICP'):
         return 'VCR|ICP'
       
  return ''

def findSelectionCriteria(x):
  for n in prod30:
    if(n["HS92"] == x):
       if(n["Criterio de selección"] == 'Proximidad y complejidad'):
         return 'proximidad|complejidad'
       if(n["Criterio de selección"] == 'Proximidad y complejidad; VCR y complejidad'):
         return 'proximidad|complejidad|VCR'
       if(n["Criterio de selección"] == 'Proximidad y complejidad; VCR 2011 e ICP.'):
         return 'proximidad|complejidad|VCR2011|ICP'
       if(n["Criterio de selección"] == 'VCR y complejidad'):
         return 'VCR|complejidad'
       if(n["Criterio de selección"] == 'VCR 2011 e ICP'):
         return 'VCR|ICP'
       if(n["Criterio de selección"] == 'VCR 2011 e ICP'):
         return 'VCR|ICP'
       
  return ''

def findEnvironmentalService(x):
    for n in prod30:
      if(n["HS92"] == x): return n["Servicio ambiental"] 
    for a in environmental_services:
        if(a["code"] == x): return a["service"]

    return ''

for x in green_products: 
  desc = findClasification(x["code"]).replace('los demas ','').replace('las demas ','').capitalize();
  if(desc): x["description"] = desc
  cat = findCategory(x["code"])
  if(cat): x["category"] = cat
  x["selection_criteria"] = findSelectionCriteria(x["code"])
  x["environmental_service"] = findEnvironmentalService(x["code"])

keys = green_products[0].keys()
with open('green_products.csv', 'w', newline='')  as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(green_products)
```
