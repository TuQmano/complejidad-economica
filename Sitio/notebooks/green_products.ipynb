```python
import pandas as pd
import numpy as np
```

```python
classif = pd.read_csv('./data/DESCRIPCION PARTIDAS Y SUBPARTIDAS - SUBPARTIDAS (6 DIGITOS).csv')
classif
```

```python
nodes = pd.read_csv('./data/informacion_nodos_2014_2018.csv')
nodes
```

```python
prod30 = pd.read_csv('./data/Tabla 30 Productos.csv')
prod30
```

```python
environmental_services = pd.read_csv('./servicios.csv')
environmental_services
```

```python
criterios = {
    'Proximidad y complejidad': 'proximidad|complejidad',
    'Proximidad y complejidad; VCR y complejidad': 'proximidad|complejidad|VCR',
    'Proximidad y complejidad; VCR 2011 e ICP.': 'proximidad|complejidad|VCR2011|ICP',
    'VCR y complejidad': 'VCR|complejidad',
    'VCR 2011 e ICP': 'VCR2011|ICP'
}
```

```python
def find(code, df, filter_column, data_column):
    result = list(df.loc[df[filter_column] == code, data_column])
    if len(result) > 0:
        result = result[0]
        return result
    return ''
```

```python
def findClassification(code):
    return find(code, classif, "SubPartidas Arancelarias", "Descripción del producto")
```

```python
def findCategory(code):
    return find(code, nodes, "prod_code", "category_description")
```

```python
def findSelectionCriteria(code):
    criterio = find(code, prod30, 'HS92', "Criterio de selección")
    return criterios.get(criterio, '')
```

```python
def findEnvironmentalService(code):
    servicio = find(code, prod30, 'HS92', "Servicio ambiental")
    if servicio == '':
        servicio = find(code, environmental_services, 'code', 'service')
    return servicio
```

```python
products = pd.DataFrame()
columns_in = ['prod_code', 'Density_ARG', 'RCA_ARG', 'RCA_BRA', 'RCA_DEU', 'RCA_CHN', 'PCI', 'COG_ARG', 'prod_description', 'Export_values_ARG', 'category_description']
columns_out = ['code', 'density', 'rca_arg', 'rca_bra', 'rca_deu', 'rca_chn', 'pci', 'cog', 'description', 'export', 'category']
products[columns_out] = nodes[columns_in]
products['id'] = products['code']

for i, node in products.iterrows():
    prod_code = node['code']
    
    description = findClassification(str(prod_code)).replace('los demas ','').replace('las demas ','').capitalize();
    if description == '':
        description = node['description'].replace(', nes', '').replace(' nes', '')

    category = node['category']
    selection_criteria = findSelectionCriteria(prod_code)
    environmental_service = findEnvironmentalService(prod_code)

    products.loc[i, 'description'] = description
    products.loc[i, 'category'] = category
    products.loc[i, 'selection_criteria'] = selection_criteria
    products.loc[i, 'environmental_service'] = environmental_service
```
