```python
import pandas as pd
import numpy as np
```

```python
df_nodos = pd.read_csv('./informacion_nodos_2014_2018.csv')
df_nodos[0:10]
```

```python
categories = [
    'Live Animals, Animal Products',
    'Vegetable Products',
    'Animal or Vegetable Fats and Oils',
    'Prepared Foodstuffs, Beverages, Tobacco',
    'Mineral Products',
    'Chemicals and Allied Industries',
    'Plastics, Rubbers',
    'Raw Hides, Skins, Leather and Furs',
    'Wood and Wood Products',
    'Paper, Pulp of Wood or of Fibrous Material',
    'Textiles',
    'Footwear, Headgear',
    'Stone, Glass, Cement, Ceramics',
    'Natural or Cultured Pearls, Precious Metals',
    'Base Metals',
    'Machinery, Electrical',
    'Transportation',
    'Precision Instruments',
    'Arms and Ammunition',
    'Miscellaneous Manufactured Articles',
    'Works of Art, Antiques',
]
```

```python
ranges = np.array([0, 5, 14, 15, 24, 27, 38, 40, 43, 46, 49, 63, 67, 70, 71, 83, 85, 89, 92, 93, 96, 97])
category_product = np.array([])
for i, top in enumerate(ranges):
    if (i != 0):
        category_product = np.append(
            category_product,
            np.full(top - ranges[i - 1], categories[i - 1])
        )
```

```python
products = pd.DataFrame()
products[['code', 'density', 'rca_arg', 'rca_bra', 'rca_deu', 'rca_chn', 'pci', 'cog', 'description', 'export']] = df_nodos[['prod_code', 'Density_ARG', 'RCA_ARG', 'RCA_BRA', 'RCA_DEU', 'RCA_CHN', 'PCI', 'COG_ARG', 'prod_description', 'Export_values_ARG']]
products['category'] = products['code'].map(lambda x: category_product[int(str(x)[0:2])])
products['id'] = products['code']
display(products.head())
products.to_csv('green_products-old.csv')
```
